```python
#Aumento de datos con ImageDataGenerator
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import numpy as np

#Crear el dataset generador #AJUSTAMOS EL TAMAO DE TODAS LAS IMAGENES
datagen = ImageDataGenerator(
    rescale=1. / 255,
    rotation_range = 30,
    width_shift_range = 0.25,
    height_shift_range = 0.25,
    shear_range = 15,
    zoom_range = [0.5, 1.5],
    validation_split=0.2 #20% para pruebas
)

#Generadores para sets de entrenamiento y pruebas
data_gen_entrenamiento = datagen.flow_from_directory('/content/dataset/', target_size=(224,224),
                                                     batch_size=32, shuffle=True, subset='training')
data_gen_pruebas = datagen.flow_from_directory('/content/dataset/', target_size=(224,224),
                                                     batch_size=32, shuffle=True, subset='validation')

#Imprimir 10 imagenes del generador de entrenamiento
for imagen, etiqueta in data_gen_entrenamiento:
  for i in range(10):
    plt.subplot(2,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(imagen[i])
  break
plt.show()
```

```python
#Link donde esta la neurona entrenada a descargar
#https://tfhub.dev/google/tf2-preview/mobilenet_v2/feature_vector/4
import tensorflow as tf
import tensorflow_hub as hub

url = "https://tfhub.dev/google/tf2-preview/mobilenet_v2/feature_vector/4"
#KerasLayer descarga el modelo de la url
mobilenetv2 = hub.KerasLayer(url, input_shape=(224,224,3))
```

```python
#Congelar el modelo descargado
mobilenetv2.trainable = False
```

```python
modelo = tf.keras.Sequential([
    mobilenetv2,
    tf.keras.layers.Dense(21, activation='softmax')
])
# 5 es la cantidad de neuronas
```

```python
modelo.summary()
```

```python
#Compilar como siempre
modelo.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
#Entrenar el modelo
EPOCAS = 50

historial = modelo.fit(
    data_gen_entrenamiento, epochs=EPOCAS, batch_size=32,
    validation_data=data_gen_pruebas
)
```

```python
#Graficas de precisin
acc = historial.history['accuracy']
val_acc = historial.history['val_accuracy']

loss = historial.history['loss']
val_loss = historial.history['val_loss']

rango_epocas = range(50)

plt.figure(figsize=(8,8))
plt.subplot(1,2,1)
plt.plot(rango_epocas, acc, label='Precisin Entrenamiento')
plt.plot(rango_epocas, val_acc, label='Precisin Pruebas')
plt.legend(loc='lower right')
plt.title('Precisin de entrenamiento y pruebas')

plt.subplot(1,2,2)
plt.plot(rango_epocas, loss, label='Prdida de entrenamiento')
plt.plot(rango_epocas, val_loss, label='Prdida de pruebas')
plt.legend(loc='upper right')
plt.title('Prdida de entrenamiento y pruebas')
plt.show()
```
